```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = "../Data/Cleaned_Fiscal.csv"  # Replace with your file path if different
data = pd.read_csv(file_path, parse_dates=["Date"], index_col="Date")

quarterly_taxes = data["Total Taxes"]

financial_crisis_start = pd.to_datetime("2007-10-01")
financial_crisis_end = pd.to_datetime("2009-06-30")
covid_crisis_start = pd.to_datetime("2020-01-01")
covid_crisis_end = pd.to_datetime("2021-12-31")

plt.figure(figsize=(14, 8))

plt.plot(quarterly_taxes.index, quarterly_taxes, marker='o', color='black', linewidth=2, label="Total Taxes")

plt.axvspan(financial_crisis_start, financial_crisis_end, color='darkgreen', alpha=0.3, label="2008 Financial Crisis")
plt.axvspan(financial_crisis_end, pd.to_datetime("2019-12-31"), color='lightgreen', alpha=0.3, label="Post-2008 Recovery")
plt.axvspan(covid_crisis_start, covid_crisis_end, color='darkgreen', alpha=0.3, label="COVID-19 Crisis")
plt.axvspan(covid_crisis_end, quarterly_taxes.index.max(), color='lightgreen', alpha=0.3, label="Post-COVID Recovery")

plt.title("Total Tax Revenue Trends with Quarterly Shading for Economic Periods", fontsize=18, weight='bold')
plt.xlabel("Quarter", fontsize=14)
plt.ylabel("Total Taxes (Billions)", fontsize=14)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=12)
plt.legend(fontsize=12, loc='upper left')
plt.grid(visible=True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
